```python
def display_all(df):
    with pd.option_context('display.max_columns', 500,'display.max_rows', 500):
        display(df)
```

```python
display_all(fifa_copy.head())
```

```python
display_all(fifa_copy.isnull().sum())
```

```python
items_to_fill = ['Crossing','Finishing','HeadingAccuracy','ShortPassing','Volleys','Dribbling','Curve','FKAccuracy','LongPassing',
                 'BallControl','Acceleration','SprintSpeed','Agility','Reactions',
       'Balance', 'ShotPower', 'Jumping', 'Stamina', 'Strength', 'LongShots',
       'Aggression', 'Interceptions', 'Positioning', 'Vision', 'Penalties',
       'Composure', 'Marking', 'StandingTackle', 'SlidingTackle', 'GKDiving',
       'GKHandling', 'GKKicking', 'GKPositioning', 'GKReflexes']
for i in items_to_fill:
    fifa_copy[i] = fifa_copy[i].fillna(fifa_copy[i].mean())
```

```python
display_all(fifa_copy.head())
```

```python
fifa_copy.drop(['Loaned From'], axis = 1, inplace = True)
```

```python
items_to_mode = ['Height', 'Weight','Preferred Foot',
       'International Reputation', 'Weak Foot', 'Skill Moves', 'Work Rate','Position','Joined','Contract Valid Until']
for i in items_to_mode:
    fifa_copy[i] = fifa_copy[i].fillna(fifa_copy[i].mode()[0])
```

```python
fifa_copy.columns
```

```python
display_all(fifa_copy.isnull().sum())
```

```python
fifa_copy['Club'] = fifa_copy['Club'].fillna('No Club')
 
```

```python
fifa_copy['Body Type'] = fifa_copy['Body Type'].fillna('No Body Type')
fifa_copy['Real Face'] = fifa_copy['Real Face'].fillna('No Real Face')
fifa_copy['Jersey Number'] = fifa_copy['Jersey Number'].fillna('No Jersey Number')
```

```python
display_all(fifa_copy.isnull().sum())
```

```python
fifa_copy.dropna(axis = 1, inplace = True)
```

```python
fifa_copy['Retirement'] = np.where(fifa_copy['Age'] <30,'No Retirement',
                                  np.where((fifa_copy['Age'] >= 30) & (fifa_copy['Age'] < 35), 'About to Retire',
                                          np.where((fifa_copy['Age'] >= 35),'Retired',None)))
```

```python
sns.boxplot(x = fifa_copy['Age_capped'])
```

```python
q1 = fifa_copy['Age'].quantile(0.25)
q3 = fifa_copy['Age'].quantile(0.75)
IQR = q3 - q1
```

```python
lower_bound = q1 - 1.5*IQR
upper_bound = q3 + 1.5*IQR
```

```python
fifa_copy['Age_capped'] = np.clip(fifa_copy['Age'], lower_bound,upper_bound)
```
